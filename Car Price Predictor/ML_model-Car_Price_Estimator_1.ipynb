# 📘 Importing Essential Libraries


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, root_mean_squared_error
```

# 📥 Loading and Inspecting the Dataset


```python
# Loading the dataset from CSV file
df = pd.DataFrame(pd.read_csv('car_price_prediction.csv'))
# Dropping the ID column as it is not useful for prediction
df.drop(['ID', 'Doors'], axis=1 , inplace=True)
df.head()
```

```python
# Getting information about dataset (nulls, dtypes, etc.)
df.info()
```

## 🧹 Data Cleaning

Cleaning `Levy`, `Mileage`, `Engine volume` .

```python

# Clean 'Levy' column
# Replacing '-' with NaN and converting 'Levy' to float
df['Levy'] = df['Levy'].replace('-', np.nan).astype(float)

# Clean 'Mileage' column
# Removing ' km' and converting 'Mileage' to integer
df['Mileage'] = df['Mileage'].str.replace(' km', '').astype(int)

# Clean 'Engine volume' column
df['Engine volume'] = df['Engine volume'].str.extract(r'(\d+\.\d+|\d+)').astype(float)
```

# 🧠 Feature Engineering

Creating a new features: `Car Age`, `Safety level` & `Safety Index` .

```python

# Add Car age as a new feature
df['Car age'] = 2025 - df['Prod. year']

# Add Safety level as a new feature
df['Airbags'] = pd.to_numeric(df['Airbags'], errors='coerce')
df['Safety level'] = df['Airbags'].apply(lambda x: 'None' if x == 0 else 'Low' if x <= 2 else 'Moderate' if x <= 6 else 'High')

# Add Safety Index as a new feature
df['Safety Index'] = (df['Airbags'] * pd.to_numeric(df['Engine volume'], errors='coerce')) / df['Car age']
df['Safety Index'].replace(np.nan, 0, inplace=True)
df
```

## 🔄 Encoding Categorical Features & Final Preprocessing


```python

# Label Encoding
#le = LabelEncoder()
#for col in df.select_dtypes(include='object').columns:
#    df[col] = le.fit_transform(df[col])

# One-hot Encoding
cat_cols = df.select_dtypes(include='object').columns
print(cat_cols)
# cat_cols = ['Manufacturer', 'Model', 'Category', 'Leather interior', 'Fuel type', 'Engine volume', 'Mileage', 'Gear box type', 'Drive wheels', 'Doors', 'Wheel', 'Color', 'Safety level']
df = pd.get_dummies(df, columns=cat_cols, drop_first=True) 

# Preprocessing

sns.histplot(df['Levy'].dropna(), kde=True, bins=30)
plt.title('Levy null value Distribution')
plt.show()

df['Levy'].skew()    # > 0    ->    right skew    =>    median > mean
# group median based on Engine Volume
df.groupby('Engine volume')['Levy'].median().dropna()   #   >500   ->   grouped median

df['Levy'] = df.groupby('Engine volume')['Levy'].transform(lambda x: x.fillna(x.median()))
df['Levy'] = df['Levy'].fillna(df['Levy'].median())
```

```python

#Managing outliers in 'Price' (target)
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df = df[(df['Price'] >= lower_bound) & (df['Price'] <= upper_bound)]
```

## 🤖 Model Training: Regression Algorithms


```python
x = df.drop('Price', axis=1)
y = df['Price']

# To split data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

lr = LinearRegression()
lr.fit(x_train, y_train)
lr_predict = lr.predict(x_test)

dt = DecisionTreeRegressor()
dt.fit(x_train, y_train)
dt_predict = dt.predict(x_test)

model = RandomForestRegressor()
model.fit(x_train, y_train)
model_predict =model.predict(x_test)
```

## 📈 Model Evaluation Metrics


```python

# Evaluating model performance for Linear Regression
print('For Linear Regression Model')
print(f'r2 score : {r2_score(y_test, lr_predict):.4f}')
print(f'MAE : {mean_absolute_error(y_test, lr_predict):.2f}')
print(f'MSE : {mean_squared_error(y_test, lr_predict):.2f}')
print(f'RMSE : {root_mean_squared_error(y_test, lr_predict):.2f}')

# Evaluating model performance for Decision Tree
print('\nFor Decision Tree Model')
r2 = round(r2_score(y_test, dt_predict), 4)
print(f'r2 score : {r2_score(y_test, dt_predict):.4f}')
mae = round(mean_absolute_error(y_test, dt_predict), 2)
print(f'MAE : {mean_absolute_error(y_test, dt_predict):.2f}')
mse = round(mean_squared_error(y_test, dt_predict), 2)            
print(f'MSE : {mean_squared_error(y_test, dt_predict):.2f}')
rmse = round(root_mean_squared_error(y_test, dt_predict), 2)
print(f'RMSE : {root_mean_squared_error(y_test, dt_predict):.2f}')
            
# Evaluating model performance for Random Forest
print('\nFor Random Forest Model')
print(f'r2 score : {r2_score(y_test, model_predict):.4f}')
print(f'MAE : {mean_absolute_error(y_test, model_predict):.2f}')
print(f'MSE : {mean_squared_error(y_test, model_predict):.2f}')
print(f'RMSE : {root_mean_squared_error(y_test, model_predict):.2f}')
```
